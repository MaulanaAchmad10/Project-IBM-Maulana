```python
# Analisis Demografi Pelanggan vs Pemesanan Makanan Online

## Pendahuluan
# Proyek ini bertujuan untuk menganalisis hubungan antara karakteristik demografi pelanggan (seperti usia, jenis kelamin, status perkawinan, pendapatan, dll.) dan kebiasaan pemesanan makanan online.
# Kita akan menggunakan dataset 'onlinefoods.csv' untuk menjawab pertanyaan-pertanyaan berikut:
# 1. Apakah ada hubungan antara demografi pelanggan dengan keputusan mereka untuk memesan makanan online ('Output')?
# 2. Bagaimana karakteristik demografi (misalnya, usia, pendapatan bulanan) memengaruhi umpan balik ('Feedback') yang diberikan oleh pelanggan?
# 3. Apa saja faktor-faktor demografi yang paling signifikan memengaruhi keputusan pemesanan ('Output')?

## 1. Persiapan Lingkungan dan Pemuatan Data

# Impor library yang diperlukan: pandas, numpy, matplotlib.pyplot, seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Atur gaya visualisasi
sns.set_style("whitegrid")
plt.style.use("seaborn-v0_8-whitegrid")

# Muat dataset 'onlinefoods.csv' ke dalam DataFrame
df = pd.read_csv('onlinefoods.csv')

## 2. Eksplorasi dan Pembersihan Data

# Tampilkan 5 baris pertama data
print("Lima baris pertama dataset:")
print(df.head())

# Tampilkan informasi ringkasan dataset (tipe data, non-null counts)
print("\nInformasi dataset:")
df.info()

# Periksa dan tangani nilai yang hilang (jika ada)
print("\nJumlah nilai yang hilang per kolom:")
print(df.isnull().sum())
# Jika ada nilai yang hilang, putuskan cara menanganinya (misalnya, mengisi dengan modus, median, atau menghapus baris)

# Bersihkan dan standarisasi nama kolom untuk kemudahan analisis
df.columns = df.columns.str.replace(' ', '_').str.lower()
print("\nNama kolom setelah dibersihkan:")
print(df.columns)

## 3. Analisis Demografi vs. Keputusan Pemesanan (Output)

# Hitung jumlah pelanggan yang memesan (Yes/No)
print("\nJumlah pemesanan:")
print(df['output'].value_counts())

# Visualisasikan hubungan antara 'Gender' dan 'Output'
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='gender', hue='output')
plt.title('Keputusan Pemesanan Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah Pelanggan')
plt.show()

# Visualisasikan hubungan antara 'Marital Status' dan 'Output'
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='marital_status', hue='output')
plt.title('Keputusan Pemesanan Berdasarkan Status Perkawinan')
plt.xlabel('Status Perkawinan')
plt.ylabel('Jumlah Pelanggan')
plt.show()

# Visualisasikan hubungan antara 'Monthly Income' dan 'Output'
plt.figure(figsize=(12, 7))
sns.countplot(data=df, x='monthly_income', hue='output', order=df['monthly_income'].value_counts().index)
plt.title('Keputusan Pemesanan Berdasarkan Pendapatan Bulanan')
plt.xlabel('Pendapatan Bulanan')
plt.ylabel('Jumlah Pelanggan')
plt.xticks(rotation=45)
plt.show()

## 4. Analisis Demografi vs. Umpan Balik (Feedback)

# Hitung jumlah umpan balik (Positive/Negative)
print("\nJumlah umpan balik:")
print(df['feedback'].value_counts())

# Visualisasikan hubungan antara 'Age' dan 'Feedback' menggunakan boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='feedback', y='age')
plt.title('Distribusi Usia Berdasarkan Umpan Balik')
plt.xlabel('Umpan Balik')
plt.ylabel('Usia')
plt.show()

# Visualisasikan hubungan antara 'Occupation' dan 'Feedback'
plt.figure(figsize=(12, 7))
sns.countplot(data=df, x='occupation', hue='feedback')
plt.title('Umpan Balik Berdasarkan Pekerjaan')
plt.xlabel('Pekerjaan')
plt.ylabel('Jumlah Pelanggan')
plt.xticks(rotation=45)
plt.show()

## 5. Kesimpulan

# Berdasarkan analisis, ringkas temuan-temuan utama.
# Contoh:
# - Kelompok demografi mana yang paling sering melakukan pemesanan?
# - Apakah ada pola pendapatan yang jelas terkait dengan keputusan pemesanan?
# - Bagaimana usia dan pekerjaan memengaruhi umpan balik pelanggan?
```

```python
# Analisis Pendapatan Pelanggan vs. Frekuensi Pemesanan

## Pendahuluan
# Proyek ini bertujuan untuk menganalisis hubungan antara tingkat pendapatan bulanan pelanggan ('Monthly Income') dan frekuensi pemesanan makanan online.
# Kita akan mengeksplorasi pertanyaan-pertanyaan berikut:
# 1. Apakah ada pola pemesanan yang berbeda di antara kelompok pendapatan yang berbeda?
# 2. Apakah pendapatan bulanan memiliki korelasi dengan keputusan pemesanan ('Output')?
# 3. Bagaimana pendapatan memengaruhi frekuensi pemesanan secara keseluruhan?

## 1. Persiapan Data
# Muat dataset dan bersihkan kolom seperti pada analisis sebelumnya.
# Pastikan nama kolom telah distandarisasi dan data siap untuk dianalisis.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Muat dataset
df = pd.read_csv('onlinefoods.csv')

# Bersihkan nama kolom
df.columns = df.columns.str.replace(' ', '_').str.lower()
df = df.rename(columns={'pin_code': 'pincode'})

# Ubah 'No Income' menjadi 0 agar bisa dianalisis secara numerik jika diperlukan, atau perlakukan sebagai kategori terpisah.
# Untuk analisis ini, kita akan memperlakukannya sebagai kategori.

## 2. Analisis Hubungan Pendapatan vs. Pemesanan

# Hitung jumlah pemesanan ('Output') per kategori pendapatan
pemesanan_per_pendapatan = df.groupby('monthly_income')['output'].value_counts().unstack().fillna(0)
print("Jumlah pemesanan per kategori pendapatan:")
print(pemesanan_per_pendapatan)

# Visualisasikan data menggunakan bar chart
# Buat bar chart yang menunjukkan jumlah 'Yes' dan 'No' untuk setiap kategori 'Monthly Income'.
pemesanan_per_pendapatan.plot(kind='bar', figsize=(12, 7))
plt.title('Keputusan Pemesanan Berdasarkan Kategori Pendapatan')
plt.xlabel('Pendapatan Bulanan')
plt.ylabel('Jumlah Pelanggan')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Memesan Makanan Online')
plt.tight_layout()
plt.show()

# Hitung persentase pemesanan 'Yes' untuk setiap kategori pendapatan
pemesanan_persentase = pemesanan_per_pendapatan.div(pemesanan_per_pendapatan.sum(axis=1), axis=0) * 100
print("\nPersentase pemesanan per kategori pendapatan:")
print(pemesanan_persentase)

# Visualisasikan persentase menggunakan stacked bar chart
pemesanan_persentase.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Persentase Keputusan Pemesanan Berdasarkan Kategori Pendapatan')
plt.xlabel('Pendapatan Bulanan')
plt.ylabel('Persentase')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Memesan Makanan Online')
plt.tight_layout()
plt.show()

## 3. Analisis Pendapatan vs. Kategori Lain

# Analisis hubungan pendapatan dengan 'Feedback'
# Buat crosstab untuk melihat hubungan antara 'Monthly Income' dan 'Feedback'.
crosstab_feedback = pd.crosstab(df['monthly_income'], df['feedback'])
print("\nHubungan Pendapatan vs. Umpan Balik:")
print(crosstab_feedback)

# Visualisasikan data
crosstab_feedback.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Umpan Balik Berdasarkan Kategori Pendapatan')
plt.xlabel('Pendapatan Bulanan')
plt.ylabel('Jumlah Pelanggan')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Umpan Balik')
plt.tight_layout()
plt.show()

## 4. Kesimpulan

# Berdasarkan analisis di atas, rumuskan kesimpulan utama.
# Contoh:
# - Kelompok pendapatan mana yang paling aktif dalam memesan makanan online?
# - Apakah ada hubungan yang jelas antara pendapatan dan jenis umpan balik yang diberikan?
# - Berdasarkan persentase, kelompok pendapatan mana yang paling mungkin melakukan pemesanan?
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Muat dataset
df = pd.read_csv('onlinefoods.csv')

# Bersihkan nama kolom untuk konsistensi
df.columns = df.columns.str.replace(' ', '_').str.lower()
df = df.rename(columns={'pin_code': 'pincode'})

# Tampilkan informasi dasar untuk verifikasi
print("Informasi Dataset:")
df.info()
print("\nLima baris pertama:")
print(df.head())
```

```python
# Hitung jumlah entri untuk setiap kategori pendidikan
print("\nDistribusi Tingkat Pendidikan:")
print(df['educational_qualifications'].value_counts())

# Hitung jumlah entri untuk setiap kategori umpan balik
print("\nDistribusi Umpan Balik:")
print(df['feedback'].value_counts())
```

```python
# Buat tabel silang antara 'educational_qualifications' dan 'feedback'
crosstab_edu_feedback = pd.crosstab(df['educational_qualifications'], df['feedback'])
print("\nTabel Silang: Pendidikan vs. Umpan Balik")
print(crosstab_edu_feedback)

# Normalisasi tabel silang untuk melihat persentase
crosstab_normalized = crosstab_edu_feedback.div(crosstab_edu_feedback.sum(axis=1), axis=0)
print("\nTabel Silang (Persentase): Pendidikan vs. Umpan Balik")
print(crosstab_normalized)
```

```python
plt.figure(figsize=(12, 7))
crosstab_normalized.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Proporsi Umpan Balik Positif vs. Negatif Berdasarkan Tingkat Pendidikan')
plt.xlabel('Tingkat Pendidikan')
plt.ylabel('Proporsi')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Umpan Balik', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Muat dataset
df = pd.read_csv('onlinefoods.csv')

# Bersihkan nama kolom
df.columns = df.columns.str.replace(' ', '_').str.lower()

# Tampilkan kategori unik dari kolom 'Occupation' untuk verifikasi
print("Kategori Pekerjaan:")
print(df['occupation'].unique())

# Buat tabel silang antara 'occupation' dan 'output'
crosstab_occ_order = pd.crosstab(df['occupation'], df['output'])
print("\nTabel Silang: Pekerjaan vs. Keputusan Pemesanan")
print(crosstab_occ_order)

# 1. Visualisasi Jumlah Pemesanan per Pekerjaan
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='occupation', hue='output', order=df['occupation'].value_counts().index)
plt.title('Keputusan Pemesanan Berdasarkan Pekerjaan')
plt.xlabel('Pekerjaan')
plt.ylabel('Jumlah Pelanggan')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Memesan Makanan Online')
plt.tight_layout()
plt.show()

# 2. Visualisasi Persentase Pemesanan per Pekerjaan (Normalisasi)
# Hitung persentase 'Yes' dan 'No' untuk setiap kategori pekerjaan
crosstab_percentage = crosstab_occ_order.div(crosstab_occ_order.sum(axis=1), axis=0) * 100

plt.figure(figsize=(10, 6))
crosstab_percentage.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Proporsi Pemesanan Berdasarkan Pekerjaan')
plt.xlabel('Pekerjaan')
plt.ylabel('Persentase')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Memesan Makanan Online')
plt.tight_layout()
plt.show()
```

```python
# Buat tabel silang antara 'family_size' dan 'output'
crosstab_family_order = pd.crosstab(df['family_size'], df['output'])
print("Tabel Silang: Ukuran Keluarga vs. Keputusan Pemesanan")
print(crosstab_family_order)

# Hitung persentase pemesanan ('Yes' dan 'No') untuk setiap ukuran keluarga
crosstab_percentage = crosstab_family_order.div(crosstab_family_order.sum(axis=1), axis=0) * 100
print("\nPersentase Pemesanan per Ukuran Keluarga:")
print(crosstab_percentage)

# Visualisasi data
plt.figure(figsize=(10, 6))
crosstab_percentage.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Proporsi Pemesanan Berdasarkan Ukuran Keluarga')
plt.xlabel('Ukuran Keluarga')
plt.ylabel('Persentase')
plt.xticks(rotation=0)
plt.legend(title='Memesan Makanan Online', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
